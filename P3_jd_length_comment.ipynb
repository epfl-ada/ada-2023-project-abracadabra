```python
import P3_helpers as hlp
import matplotlib.pyplot as plt
```

```python
df = hlp.get_dataframe()
```

```python
df['Parsed_Comment'] = df['Comment'].apply(hlp.get_parsed_comment)
df = df[~df.Parsed_Comment.isnull()]
df['Comment_Length'] = df['Parsed_Comment'].apply(lambda x: len(x))
```

```python
mean_charcter_length = df['Comment_Length'].mean()
median_charcter_length = df['Comment_Length'].median()

print('Mean character length: {}'.format(mean_charcter_length))
print('Max character length: {}'.format(df['Comment_Length'].max()))
print('Median character length: {}'.format(median_charcter_length))
```

```python
#Plot the distribution of comment lengths
plt.hist(df['Comment_Length'], bins=1000)
plt.title('Distribution of Comment Lengths (in words)')
plt.xlabel('Comment Length - log scale')
plt.ylabel('Number of Comments')
plt.xscale('log')
plt.show()
```
