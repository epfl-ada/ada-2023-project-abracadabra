```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.tools as tls

import numpy as np
```

```python
# generating data
x = np.linspace(0, 10, 1000)
y = np.sin(x)

# plotting
fig = plt.figure(figsize=(8,6))
sns.set_style("whitegrid")
plt.plot(x, y, color="green", lw=3)
plt.xlim(0, 10)
plt.ylim(-1.2, 1.2)
plt.xlabel("x", fontsize=16)
plt.ylabel("sin(x)", fontsize=16)
plt.show()

# convert to plotly
py_plot = tls.mpl_to_plotly(fig)

# save in html
py_plot.write_html("../assets/plot/sin.html")
```
